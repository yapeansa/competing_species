```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.close("all")
```

```python
fixed_points = np.array([[0., 0.], [0., 3.], [1., 1.], [2., 0.]])
```

```python
x = np.linspace(0, 4, 10)
y = np.linspace(0, 4, 10)
X, Y = np.meshgrid(x, y)
```

```python
Xdot = X * (2 - X - Y)
Ydot = Y * (3 - 2 * X - Y)
```

```python
# nullclines
x_nullcline1 = np.linspace(0, 4, 10)  # x-nullcline: y = 2 - x
y_nullcline1 = np.linspace(0, 4, 10)  # y-nullcline: y = 3 - 2x
```

```python
fig, ax = plt.subplots()
ax.streamplot(X, Y, Xdot, Ydot, color="blue", density=1.2)
ax.scatter(*fixed_points.T, color="red")
```

```python
# Add nullclines
ax.plot(x_nullcline1, 2 - x_nullcline1, "orange", label="x-nullcline")
ax.plot(x_nullcline1, 3 - 2 * x_nullcline1, "green", label="y-nullcline")
ax.axhline(0, color="g", linestyle="--", alpha=0.7)  # y = 0 nullcline
ax.axvline(0, color="g", linestyle="--", alpha=0.7)  # x = 0 nullcline
```

```python
# Labels and legend
ax.set_xlim(-0.125, 4)
ax.set_ylim(-0.125, 4)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.title("Phase Portrait with Nullclines")
# plt.show()
```
